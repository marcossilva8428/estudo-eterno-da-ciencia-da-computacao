Variáveis e instruções

- Variável é um nome que se refere a um valor
- Usamos uma instrução de atribuição.

```python
n = 17
```

- A instrução de atribuição tem 3 partes, nome da variávela esquerda, operador de igualdade '=' e uma expressão a direita.

- Quando executamos uma instrução de atribuição não existe saída, mas é possível utilizar a variável como uma expressãopara exibir o valor.

Expressões e instruções

Exemplo de expressão 

```python
19 + n + round(math.pi) * 2
```

- Instrução é uma unidade de código que tem um efeito mas nenhum valor.

```python
n = 19
```

- Calcular o valor de uma expressão é o que chamamos de avaliação. Executar uma instrução é chamado de execução.
