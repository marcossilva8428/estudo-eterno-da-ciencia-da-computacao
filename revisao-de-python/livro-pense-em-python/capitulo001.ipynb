

Funções aritméticas

Alem dos operadores aritmeticos , o python oferece diversas funções que operam com números.
- Função round recebe um número de ponto flutuante e o arredonda para o número inteiro mais próximo.


```python
round(42.4)
```

```python
round(42.6)
```

- A função abs calcula o valor absoluto de um número, em caso de um número positivo, o valor absoluto é o próprio número.

```python
abs(42)
```

- Em caso de número negativo, o valor absoluto é o valor positivo correspondente.

```python
abs(-42)
```

- Quando utilizarmos a função dessa forma, estamos chamando a função . Uma expressão que chama uma função é chamada de chamada de função.
- Ao chamar uma função, é obrigatório utilizar parênteses caso você omita, uma mensgem de erro será exibita.

```python
abs 42
```

- Este erro indica que a estrutura da expressão está incorreta. Neste exeplo , é falta de parênteses.
- Se omitir o parênteses e o valor, é uma expressão valida que contem um valor.

```python
abs
```

Strings 

- Sequência de letras, conhecidas como strings porque as letras são encadeadas como contas em um colar. Para definir uma string, podemos colocar uma sequência de letras entre aspas simples :

'Hello'

- Também é permitido aspas duplas.

"Wold"

- É útil usar aspas duplas quando precisamos incluir apostofro.

" It's a small'

- O perador "+" funciona com string. Com ele fazemos concatenação.

```python
string = 'Well,' + "It's a small " + 'world.'
print(string)
```

- O operador * multiplica as strings.

```python
string = 'spam,' * 4
print(string)
```

- Outros operadores não funcionam.
- Função len(), que calcula o comprimento de uma string.

```python
len('Spam')
```

- O len contabiliza só as letras e não as aspas.
