- A função abs calcula o valor absoluto de um número, em caso de um número positivo, o valor absoluto é o próprio número.

Strings 

- Sequência de letras, conhecidas como strings porque as letras são encadeadas como contas em um colar. Para definir uma string, podemos colocar uma sequência de letras entre aspas simples :

'Hello'

- Também é permitido aspas duplas.

"Wold"

- É útil usar aspas duplas quando precisamos incluir apostofro.

" It's a small'

- O perador "+" funciona com string. Com ele fazemos concatenação.

```python
string = 'Well,' + "It's a small " + 'world.'
print(string)
```

- O operador * multiplica as strings.

```python
string = 'spam,' * 4
print(string)
```

- Outros operadores não funcionam.
- Função len(), que calcula o comprimento de uma string.

```python
len('Spam')
```

- O len contabiliza só as letras e não as aspas.

Valores e tipos

- 2 é um intero
- 42.0 é um número de ponto flutuante e 'Hello' é uma string.

- Um tipo de valor é conhecido como tipo.
- Todo valor tem um tipo - ou , em algumas ocasiões, dizemos que ele 'pertence' a um tipo especifico.
- O Python disponibiliza uma função chamada type que indica o tipo de qualquer valor. Um valor inteiro tem o tipo int.

```python
type(2)
```

- Um número decimal, ou de ponto flutuante, tem o tipo float.

```python
type(42.0)
```

- E uma sequência de caracteres, ou string, tem tipo str

```python
type('Hello, world!')
```

- Os tipos int, float e str podem ser utilizados como funções.

```python
int(42.9)
```

```python
float(42)
```

- Se escrevermos um numero entre aspas, na verdade ele é uma string.

```python
type('126')
```

- Se tentar usar um operador aritimetico entre um str e um int, ocorre erro.

```python
'126' / 3
```

- Esse exemplo gera um TypeError, o que significa que os valores na expressão, os chamados operandos, têm tipos incompativeis.
- O operador '/' não pode ser aplicado a esse tipo de valores, que são entre str e int.
- É possível usar a função int() para converter uma string em inteiro

```python
int('126') / 3
```

- Da mesma forma com float

```python
float('12.6')
```

- Ao digitar números inteiros grandes pode ser que o python interprete como uma sequência de números inteiros separados por vígulas. Ex: notação americana.

```python
numero = 1,000,000 # notação americana seria 1.000.000
print(numero)
```

- Para tornar números grandes mais legíveis, usar underscore.

```python
numero = 1_000_000
print(numero)
```

- Quantas milhas existem em 10 quilômetros? Dica: uma milha é equivalente a 1,61 quilômetro.

```python
quantas_milhas = 10 / 1.61
print(quantas_milhas)
```

- Se você correr 10 quilômetros em 42 minutos e 42 segundos, qual será o seu passo médio(tempo por milha em minutos e segundos)?

```python
round(minutos / quantas_milhas, 2)
```
