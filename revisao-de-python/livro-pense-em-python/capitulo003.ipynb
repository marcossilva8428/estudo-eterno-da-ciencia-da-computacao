Funções

- Definindo novas funções, uma definição de uma nova função e a sequência de instruções que serão executadas quando essa função for chamada.
- def é a palavra reservada que indica uma definição de função.
- a primeira linha é chamada de cabeçalho e o restante é chamado de corpo.
- definir uma função cria um objeto de função.

```python
def imprimir(a):
    print(f'Hello, {a}!')
```
