Funções

- Definindo novas funções, uma definição de uma nova função e a sequência de instruções que serão executadas quando essa função for chamada.
- def é a palavra reservada que indica uma definição de função.
- a primeira linha é chamada de cabeçalho e o restante é chamado de corpo.
- definir uma função cria um objeto de função.

```python
def imprimir(a):
    print(f'Hello, {a}!')
```

```python
imprimir
```

- Funções podem trabalhar em conjunto

Repetição

```python
for i in range(2):
    print(i)
```

```python
range
```

- a primeira linha do for é o cabeçalho.
- a segunda é o corpo que precisa ser identado.
- a primeira linha começa com a palavra reservada for seguida de uma variável i e outra palavra reservada in. Ela utiliza a função range para criar uma sequência de dois valores. Ex: range(2)
- ao executar a instrução for o primeiro valor gerado por range é atribuido a i.
- quando chega ao final do corpo, o loop retorna para o cabeçalho, e é por isso que essa instrução é chamada de loop.
- mesmo que não utilizemos o i no corpo do loop, ainda sim é necessário declarar um nome de variável no cabeçalho.
- variáveis e parâmetros são locais.

Tracebacks

- Quando ocorre um erro de tempo de execução em uma função. o python exibe o nome da função que estava sendo executada, seguido pelo nome da função que a chamou . 

```python
line1 = 'ArithmeticError'
line2 = 'ZeroDivisionError'
```

```python
def print_twice(string):
    print(cat)
    print(cat)
```

```python
def cat_twice(part1, part2):
    cat = part1 + ' ' + part2
    print_twice(cat)
```

```python
cat_twice(line1, line2)
```

- A mensagem de erro inclui um traceback, que revela a função que estava sendo executada quando ocorreu o erro, a função que chamou e assim por diante. Neste exemplo, o traceback indica que cat_twice chamou print_twice e o erro aconteceu dentro de print_twice.
- A ordem das funções no traceback é a mesma que a ordem dos frames no diagrama de pilha.
